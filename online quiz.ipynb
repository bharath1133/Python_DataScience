```python
import random

def start_quiz():
    positive_score=0
    negative_score=0
    questions_count=0
    correct_answers=0
    wrong_answers=0
    print("Welcome to the online quiz {name}.\033[1mYour quiz consists of 5 questions. To win, get atleast 2.5 marks.\033[0m".format(name=input("Enter your name to start the quiz:")))
    print('''\033[1mPlease follow below quiz rules.\033[0m
    For every correct answer 1 mark,
    For every wrong answer 1/4 is deducted,
    If you want you can skip the question and move to next question''')
    while True:
        if len(questions)>4:
            random.shuffle(questions)
            questions_count=questions_count+1
            print("\033[1m" + questions[0][0] + "\033[0m")
            for i in range(1,4):
                  print(questions[0][i])
        else:
            print("Sorry,for the inconvienience,As of now we don't have possibe number of questions.Please try again later.Thank you.")
            break
    
    
        user_answer=input("Answer:")
        if user_answer.upper()==questions[0][4]:
            print("Your answer is correct")
            positive_score= positive_score+1
#             print("positive score:",positive_score)
#             print("negative_score:",negative_score)
            correct_answers=correct_answers+1
        elif user_answer.lower()=="skip":
            questions_count=questions_count-1
        else:
            print("Youer answer is wrong")
            wrong_answers=wrong_answers+1
            negative_score=negative_score+1/4
            positive_score= positive_score-1/4
#             print("positive score:",positive_score)
#             print("negative_score:",negative_score)
                
        if questions_count>4:
            print("You have successfully completed the quiz.Thank you.Below are your score details and status.")
            print("total correct answers:",correct_answers)
            print("tota wrong answers:",wrong_answers)
            if positive_score>=2.5:
                print("Your total score is:",positive_score)
                print("congratulations..You won the quiz,Thank you keep it up.")
                break
            else:
                print("Your total score is:",positive_score)
                print("You loose the quiz,Try again to win the quiz.Thank you.")
                break

start_quiz()
while True:
    start_quiz_again=input("If you want to start the quiz again please type yes:")
    if start_quiz_again.lower()=="yes":
        start_quiz()
    else:
        print("Thank you,Have a nice day.")
        break
```
